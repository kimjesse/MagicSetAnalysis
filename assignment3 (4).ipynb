# Visualization Technique:

As stated above I have four analyses that I wish to include in my dashboard; the distribution of market prices of cards, the distribution of "color identities", the distribution of the card popularity rankings from EDHREC, and the number of words dedicated to rules text versus the number of words dedicated to "flavor" text. I shall now break down how I will approach each of these analyses:

## The Distribution of Market Prices of Cards as Histogram:

Cards in magic are sold on the secondary market on sites such as TCGplayer and Card Kingdom. The prices of cards are often dependent on how recently the card was printed or reprinted and how popular the card is for play, the supply and demand. Other times, factors such as popularity of specific arts or other factors can increase the price on certain cards. This visualization is to explore the distribution of the prices of cards according to TCGplayer. As I am looking for the frequency and distribution of numeric data (price), I chose to depict this in a histogram with price on the x-axis and frequency on the y-axis. A histogram is a good way to graphically represent frequency and distribution of data when your data is continuous and numeric. It does this by placing values within 'bins' of a discrete range and creating a bar from the bin based on how many values fell within that range. By showing the data in this way it is easy to see how and where the data fell in the distribution. The bin sizes and number of bins can be adjusted based on how you want to break down the data for your distribution. For my dashboard I will also be adding interactivity elements such as seeing bin range and number of values in a bar/bin on mouseover and ways to filter the dashboard as a whole by set and by card types.

## The Distribution of "Color Identities" as Pie Chart:

One particular format, a way to play Magic, that I play often is called "Commander" or "EDH" (Elder Dragon Highlander). Within commander, one deck building restriction is that your deck must only contain cards that are within your commander's "[color identity](https://mtg.wiki/page/Color_identity)". I wanted to see which color identities were more represented in each set. Because color identity is a categorical variable, cards can only belong to one color identity, and all cards fall into a color identity, I felt that a pie chart would be a good way to display this information. In a pie chart, the proportions of each category are displayed as wedges of a single circle or 'pie'. By showing the data this way, it is easy for the viewer to visually see the relative proportions of each of the groups. A pie chart can be used when you want to quickly and easily display the relative proportions of groups within a whole. Pie charts can be a poor choice if the group proportions do not add up to the whole. They can also be a poor choice if being used in direct one to one comparisons of proportions across two pie charts, as it is hard to visually notice the differences in proportions when they are minor. However, I felt that a pie chart was still appropriate in my use case, as I am only interested in getting a quick visual inspection of the breakdowns and less in the set to set comparison. Also, this issue is alleviated by the fact that I will be adding the names and percentages as labels to my pie chart. Moreover, I will be adding the aforementioned filters on set and card type, as well as a mouseover label that gives the number of cards in that category.

## The Distribution of Card Popularity Rankings from EDHREC as Box and Whisker Plot

EDHREC is a popular website used by many commander players that collates data gathered from decks that commander players have posted. One thing that they provide is a ranking for each card in Magic based on how many times the card was used in a deck versus all decks. There are certain cards that are able to fit into most decks due to a variety of factors, others that only get played in a couple of decks. These EDH ranking were supplied from the Scryfall dataset and I wanted to see which sets had cards that were played more and also see how play popularity was distributed across the different rarities of cards. Given that rarity is a categorical variable and that ranking was numeric, I felt that a box and whisker plot would be a good way to visualize how the cards in the set were distributed. In a box and whisker plot the min, max, median, first, and third quartile are first plotted. The first quartile to the third quartile, the interquartile range, is displayed in a 'box' with a line at the median. Then there are lines to the min and max as 'whiskers'. Additionally, you can plot multiple box and whiskers next to each other with another variable. This way of visualizing the distribution allows the reader to quickly get a sense of the distribution of the data. Also, the addition of breaking it down by rarity and putting them next to each other makes it easy to compare across groups. Additionally, I will be adding the aforementioned filters on set and card type, along with a mouseover label that shows the previously mentioned statistics.

## The Number of Words Dedicated to Rules vs. Flavor As Scatterplot:

On a Magic card there is usually text describing what the card does when played or other actions are taken. Additionally there is sometimes text that is not used for describing any rules, often called "flavor text" which explores some background information about the character(s) or world on the card or sometimes just witty jokes. One criticism that sometimes gets raised is that cards no longer have good or interesting flavor text anymore. As cards have gotten more mechanically complex over the years, this led to less space being available for flavor text and sometimes less focus on developing the world. As someone who enjoys flavor text, I wanted to see this progression across sets for myself. To achieve this, I plotted the length of the "oracle" (numeric) or rules text on the x-axis and the length of the flavor text (numeric) on the y-axis on a scatter plot. A scatter plot allows you to visualize the relationship between two different variables, often numeric, and see any trends or patterns between the two variables. I will also be adding the aforementioned filters on set and card type, along with a mouseover label that shows length of both the oracle text and flavor text, the rarity, and the name of the card.

## Cleaning and Formatting the Data

#### Some Date formatting and filtering

I noticed that some set release dates from the wiki didn't contain days due to them being announced for the future. In order to still use this column, I converted the column to datetime and had pandas figure out the formatting and use the first day of the month as the release date for those sets.

Additionally, there are cards that were in the Scryfall database which are either online only cards for MTG Arena and MTG Online, or cards with very limited printings that aren't circulated. I wanted to exclude these card from my analysis. 

```python
#Converting the Release column to be a datetime, some sets only had a month so those will default to the first of the month
set_dat['Released'] = pd.to_datetime(set_dat['Released'],format='mixed') 
#Dropping anything that's not available on the biggest online retailer for magic cards
card_dat = card_dat[card_dat['tcgplayer_id'].notnull()] 
#Those are likely to be online only cards or other reasons that are safe to ignore
```

#### Adding Color Identity Groups

Because I wanted to visualize the breakdown of the color identities in each set, I needed a column I could build my visualization on. The Scryfall data contained a column for color identity, but it was built as a list of colors in the identity. I wanted a clearer group that I could group around so I made a column that translates the list of colors into the name for that color pairing within Magic.

```python
#Color Identity is currently a list of colors, adding a column that gives their 'lore' name
color_combos = list(card_dat['color_identity'].value_counts().index) #Finding the list of unique color combos
color_combos=["".join(id) for id in color_combos] #Turning the list into just a string so I can map it later
#Creating a list of color combo names in magic in the order that they were found in color_combos
color_names = [ 
    'Red','White','Black','Blue','Green','Colorless', 
    'Rakdos','Azorius','Dimir','Boros','Selesnya','Gruul','Izzet','Simic','Golgari','Orzhov', 
    'Grixis','Naya','Five-Color','Bant','Esper','Jeskai','Jund','Mardu','Sultai','Temur','Abzan', 
    'Altruism','Growth','Agression','Artifice','Chaos'  
]
id_col = dict(zip(color_combos,color_names)) #Zipping the two list to make a dictionary I can map to

card_dat['identity_name'] =  card_dat['color_identity'].map(lambda x: "".join(x)).map(id_col) #Now I have column I can reference for the name
```

#### Adding Card Type columns

I also wanted the ability for the users to select which card types they wanted to view. The original dataset had the type line for the cards, but the type line also contains things like the subtypes of the cards. I wanted a more explicit column that I could just pull from.

```python
#Adding some columns that give the 'type' of the card
#Some cards have multiple types, therefore I'm making each it's own column
card_dat['is_creature'] = card_dat['type_line'].str.contains(r'Creature')
card_dat['is_instant'] = card_dat['type_line'].str.contains(r'Instant')
card_dat['is_sorcery'] = card_dat['type_line'].str.contains(r'Sorcery')
card_dat['is_artifact'] = card_dat['type_line'].str.contains(r'Artifact')
card_dat['is_enchantment'] = card_dat['type_line'].str.contains(r'Enchantment')
card_dat['is_battle'] = card_dat['type_line'].str.contains(r'Battle')
card_dat['is_legendary'] = card_dat['type_line'].str.contains(r'Legendary')
card_dat['is_walker'] = card_dat['type_line'].str.contains(r'Planeswalker')
```

#### Fixing Set Names between the two Datasets

The two datasets I'm using have some slightly different names for some sets. I first wanted to see which sets weren't joining from the MTG Wiki.

```python
first_merge = card_dat.merge(set_dat,how='right',left_on='set_name',right_on='Set')
first_merge[first_merge['set_name'].isna()]['Set'].unique()
```

That's a lot of sets but a good deal of them are actually "Secret Lair" sets which Scryfall all lumps into just "Secret Lair Drop". There are other small differences between the two datasets that I will all fix with Regex. Also saving a list of the ordered sets for use later when I am building the set name dropdown.

```python
#The mtg wiki and scryfall used some different naming schemes for the sets
#changing the names from the wiki to match with scryfall
set_dat['Set'] = set_dat['Set'].str.replace(r"(?<=Commander)\s(Decks?|decks)","", regex=True)
set_dat['Set'] = set_dat['Set'].str.replace(r"Secret Lair Drop .*","Secret Lair Drop", regex=True)
set_dat['Set'] = set_dat['Set'].str.replace(r"Dungeons & Dragons:\s","", regex=True)
set_dat['Set'] = set_dat['Set'].str.replace(r"FINAL FANTASY Commander","Final Fantasy Commander", regex=True) 
set_dat['Set'] = set_dat['Set'].str.replace(r"Doctor Who Commander","Doctor Who", regex=True) 
set_dat['Set'] = set_dat['Set'].str.replace(r"Global Series: Jiang Yanggu & Mu Yanling","Global Series Jiang Yanggu & Mu Yanling", regex=True)
set_dat['Set'] = set_dat['Set'].str.replace(r"Magic: The Gathering—FINAL FANTASY","Final Fantasy", regex=True) 
set_dat['Set'] = set_dat['Set'].str.replace(r"Kamigawa: Neon Dynasty Commander","Neon Dynasty Commander", regex=True) 
set_dat['Set'] = set_dat['Set'].str.replace(r"Streets of New Capenna Commander","New Capenna Commander", regex=True) 
set_dat['Set'] = set_dat['Set'].str.replace(r"Innistrad: Midnight Hunt Commander","Midnight Hunt Commander", regex=True)
set_dat['Set'] = set_dat['Set'].str.replace(r"Innistrad: Crimson Vow Commander","Crimson Vow Commander", regex=True)
set_dat['Set'] = set_dat['Set'].str.replace(r"Adventures in the Forgotten Realms Commander","Forgotten Realms Commander", regex=True)
set_dat['Set'] = set_dat['Set'].str.replace(r"Magic: The Gathering Foundations","Foundations", regex=True) 
set_dat['Set'] = set_dat['Set'].str.replace(r"Middle-Earth","Middle-earth", regex=True) 
set_dat['Set'] = set_dat['Set'].str.replace(r"The Lord of the Rings: Tales of Middle-earth Commander","Tales of Middle-earth Commander", regex=True) 
set_dat['Set'] = set_dat['Set'].str.replace(r"Modern Horizons II\b","Modern Horizons 2", regex=True) 
set_dat['Set'] = set_dat['Set'].str.replace(r"Modern Horizons III\b","Modern Horizons 3", regex=True) 
set_dat['Set'] = set_dat['Set'].str.replace(r"Sixth Edition","Classic Sixth Edition", regex=True) 
set_dat['Set'] = set_dat['Set'].str.replace(r"Alpha \(Limited Edition\)","Limited Edition Alpha", regex=True) 
set_dat['Set'] = set_dat['Set'].str.replace(r"Beta \(Limited Edition\)","Limited Edition Beta", regex=True) 
set_order = set_dat['Set'].unique() #Saving a list of the order of the sets
```

```python
#Initially doing a left merge to see what cards/sets aren't merging
name_cleaning_card_dat = card_dat.merge(set_dat,how='right',left_on='set_name',right_on='Set')
#Getting the rows which didn't have a set in set_dat
to_fix = name_cleaning_card_dat[name_cleaning_card_dat['set_name'].isna()]
to_fix['Set'].unique()
#Magic will sometimes create gamepieces such as tokens or other cards which aren't 'normal' cards that are still in the database.
#Between this and the previous cell, the objective is to make sure every 'regular' set is at least joining.
#Building out a list of sets to exclude to make it easier to see what's left
excluded_sets = r"Event Deck|Part|Cube|Anthology|Alchemy|Secret Lair|sheet"
#see which sets are still left to fix
to_fix[~to_fix['Set'].str.contains(excluded_sets)]['Set'].unique()
#All that's remaining looks like small promo sets or specific one off printings, at this point I'm confident I covered all the normal sets. 
```

I left some sets that didn't join, but these are for sets that haven't released yet or just "sealed" sets which contain cards that are also in other "normal sets". I'm comfortable with just proceeding.

## Putting Them Together in a Plotting Function

Now I'm happy with my individual visualizations, let's build them inside another function that I can call with the interactive dashboard. It's going to take the filtered dataset after all the filters from the Ipython Widgets have been applied. Once this function is called, it should automatically build all my visualizations together.

```python
def create_plots(data):
    fig = make_subplots(rows=2, cols=2,specs=[[{'type':'xy'},{'type':'domain'}],[{'type':'xy'},{'type':'xy'}]],
                       subplot_titles=("Distribution of Prices in Set", "Breakdown of Color Identities in Set", "EDHREC Rank Distribution in Set", "Rules Text vs Flavor Text on Cards in Set"))
    
    x_val = data['prices'].str.get('usd').astype('float') #Pull the USD prices for each card
    hist = px.histogram(x=x_val) 
    fig.update_xaxes(title="Price in $", row=1, col=1,type='linear',autorange=True)
    fig.update_yaxes(title="Number of Cards with Log Scale",type='log',autorange=True,row=1, col=1)
    pie_df = data.groupby('identity_name').size().to_frame().reset_index()
    pie = px.pie(pie_df, values=0, names='identity_name')
    pie.update_traces(textinfo='label+percent', textposition='inside',
              texttemplate='<b>%{label}</b><br>%{percent:.1%}') # Custom template for bold label and one decimal percent 
    
    boxplot = px.box(data,x='rarity',y='edhrec_rank')
    fig.update_xaxes(categoryorder='array', categoryarray=["common", "uncommon", "rare","mythic"],title="Rarity", row=2, col=1)
    fig.update_yaxes(title="EDHREC Rank in Log Scale",autorange='reversed',type='log', row=2, col=1)

    scatter_plot = px.scatter(data[data['layout']=='normal'],
                              x='oracle_len',y='flavor_len',color='rarity',
                              category_orders={"rarity": ["common", "uncommon", "rare","mythic"]},hover_data=['name'])

    fig.update_layout(legend=dict(
        yanchor="top",
        y=0.35,
    ))
    fig.update_xaxes(title="Number of words in rules text", row=2, col=2)
    fig.update_yaxes(title="Number of words in flavor text", row=2, col=2)
    
    for trace in hist.data:
        fig.add_trace(trace, row=1, col=1)
    for trace in pie.data:
        fig.add_trace(trace, row=1, col=2)
    for trace in boxplot.data:
        fig.add_trace(trace, row=2, col=1)
    for trace in scatter_plot.data:
        fig.add_trace(trace, row=2, col=2)
    fig.update_layout(uniformtext_minsize=12,height=650,width=1400,title={'text':'Analysis of Sets in Magic the Gathering', 'y':0.95,'x':0.5,'xanchor': 'center','yanchor': 'top'})
    fig.data[1].showlegend = False
    fig.show()
    
create_plots(magic_dat[magic_dat['set_name'] == 'Ixalan']) #I'm always going to filter by a set so filtering here also
```

Tweaked some values so all the visualizations look nice when they are put next to each other and there aren't any competing features. Also added the titles for the axes and the overall title for the visualization. 

## Adding the Interactivity to the Visualization

Now that I have my plotting function built, I can build in the interactivity using Ipython Widgets. I'm going to add buttons that the user can check whether they want to include certain card types in their analysis. I will also need to add a dropdown menu that contains all the sets in Magic so that the user can select which set they want to view. I will then use the "interactive" function of Ipython in order to handle the calling and processing the events from the widgets in the dashboard. These will be processed in the update function which automatically update the parameters of the data that the viewer wants to see and call the plotting function again. 

```python
def create_magic_dash():

    #Generating the list of sets that the user can select from, in the order that they were released in so that it is easier to find sets.
    set_names = [set_name for set_name in set_order if set_name in magic_dat['set_name'].unique()] 
    set_dropdown = widgets.Dropdown(
        options=set_names,        # List of set names
        value='Return to Ravnica',                 
        description='Set:',     
        style={'description_width': 'initial'}  # Make label width dynamic
    )
    
    #Making a button for each card type for the user to select cards.
    check_creature = widgets.Checkbox(
        value=True,
        description='Creature',
        disabled=False,
        indent=False
    )
    check_sorcery = widgets.Checkbox(
        value=True,
        description='Sorcery',
        disabled=False,
        indent=False
    )
    check_instant = widgets.Checkbox(
        value=True,
        description='Instant',
        disabled=False,
        indent=False
    )
    check_enchantment = widgets.Checkbox(
        value=True,
        description='Enchantment',
        disabled=False,
        indent=False
    )
    check_artifact = widgets.Checkbox(
        value=True,
        description='Artifact',
        disabled=False,
        indent=False
    )
    check_battle = widgets.Checkbox(
        value=True,
        description='Battle',
        disabled=False,
        indent=False
    )
    check_walker = widgets.Checkbox(
        value=True,
        description='Planeswalker',
        disabled=False,
        indent=False
    )

    #Update Function that will be called whenever the user interacts with the dashboard. Updates the current values of the filters being used and then calls the plotting function.
    def update(creature=True,sorcery=True,instant=True,enchantment=True,artifact=True,battle=True,walker=True,set_selection='Return to Ravnica'):
        filtered_dat = magic_dat[
                            (((magic_dat['is_creature'] == True) & creature) |
                            ((magic_dat['is_sorcery'] == True) & sorcery) |
                            ((magic_dat['is_instant'] == True) & instant) |
                            ((magic_dat['is_enchantment'] == True) & enchantment) |
                            ((magic_dat['is_artifact'] == True) & artifact) |
                            ((magic_dat['is_battle'] == True) & battle) |
                            ((magic_dat['is_walker'] == True) & walker)) &
                            (magic_dat['set_name'] == set_selection)
                        ]
        clear_output(wait=True)
        create_plots(filtered_dat)

    #Using the interactive function to handle all the Ipython widgets and updating the values
    widgetbox = interactive(
        update,
        creature = check_creature,
        sorcery = check_sorcery,
        instant = check_instant,
        enchantment = check_enchantment,
        artifact = check_artifact,
        battle = check_battle,
        walker = check_walker,
        set_selection=set_dropdown
    )
    #Tweaking it so that the ipython widgets are neatly organized in rows rather than being in one long vertical line.
    controls = HBox(widgetbox.children[:-1], layout = Layout(flex_flow='row wrap'))
    output = widgetbox.children[-1]
    display(VBox([controls, output]))

create_magic_dash()
```
